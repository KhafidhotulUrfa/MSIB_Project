## Menggabungkan Dataset

```python
frames = [load_bmhs, load_dvla, load_heal, load_inaf, load_kaef, load_klbf, load_mika, load_mtmh,
          load_omed, load_peha, load_prda, load_rsgk, load_same, load_sido, load_silo, load_soho,
          load_sraj, load_tspc]
result = pd.concat(frames).drop_duplicates().reset_index(drop=True)
```

```python
# Export to Excel
from google.colab import files

xlsx = result.to_excel('dataset_saham_kesehatan1.xlsx')
files.download('dataset_saham_kesehatan1.xlsx')
```

## Menambahkan Kolom Return

```python
result['Return'] = result.groupby('Kode Saham')['Close'].pct_change()

result.head()
```

## Mengisi Missing Value Pada Return

```python
result['Return'].isnull().sum()
```

```python
result['Return'].fillna(0, inplace=True)

result['Return'].isnull().sum()
```

## Menghapus Data yang Bernilai Null

```python
result.isnull().sum()
```

```python
result = result.dropna()
```

```python
result.isnull().sum()
```

## Menambahkan Kolom Moving Average

```python
result['MA'] = result.groupby('Kode Saham')['Close'].transform(lambda x: x.rolling(3, 1).mean())

result
```
